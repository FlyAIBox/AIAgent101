```python
# ğŸ” ç¯å¢ƒä¿¡æ¯æ£€æŸ¥è„šæœ¬
#
# æœ¬è„šæœ¬çš„ä½œç”¨ï¼š
# 1. å®‰è£… pandas åº“ç”¨äºæ•°æ®è¡¨æ ¼å±•ç¤º
# 2. æ£€æŸ¥ç³»ç»Ÿçš„å„é¡¹é…ç½®ä¿¡æ¯
# 3. ç”Ÿæˆè¯¦ç»†çš„ç¯å¢ƒæŠ¥å‘Šè¡¨æ ¼
#
# å¯¹äºåˆå­¦è€…æ¥è¯´ï¼Œè¿™ä¸ªæ­¥éª¤å¸®åŠ©æ‚¨ï¼š
# - äº†è§£å½“å‰è¿è¡Œç¯å¢ƒçš„ç¡¬ä»¶é…ç½®
# - ç¡®è®¤æ˜¯å¦æ»¡è¶³æ¨¡å‹è¿è¡Œçš„æœ€ä½è¦æ±‚
# - å­¦ä¹ å¦‚ä½•é€šè¿‡ä»£ç è·å–ç³»ç»Ÿä¿¡æ¯

# å®‰è£… pandas åº“ - ç”¨äºåˆ›å»ºå’Œå±•ç¤ºæ•°æ®è¡¨æ ¼
# pandas æ˜¯ Python ä¸­æœ€æµè¡Œçš„æ•°æ®å¤„ç†å’Œåˆ†æåº“
%pip install pandas==2.2.2 tabulate==0.9.0

import platform # å¯¼å…¥ platform æ¨¡å—ä»¥è·å–ç³»ç»Ÿä¿¡æ¯
import os # å¯¼å…¥ os æ¨¡å—ä»¥ä¸æ“ä½œç³»ç»Ÿäº¤äº’
import subprocess # å¯¼å…¥ subprocess æ¨¡å—ä»¥è¿è¡Œå¤–éƒ¨å‘½ä»¤
import pandas as pd # å¯¼å…¥ pandas æ¨¡å—ï¼Œé€šå¸¸ç”¨äºæ•°æ®å¤„ç†ï¼Œè¿™é‡Œç”¨äºåˆ›å»ºè¡¨æ ¼
import shutil # å¯¼å…¥ shutil æ¨¡å—ä»¥è·å–ç£ç›˜ç©ºé—´ä¿¡æ¯

# è·å– CPU ä¿¡æ¯çš„å‡½æ•°ï¼ŒåŒ…æ‹¬æ ¸å¿ƒæ•°é‡
def get_cpu_info():
    cpu_info = "" # åˆå§‹åŒ– CPU ä¿¡æ¯å­—ç¬¦ä¸²
    physical_cores = "N/A"
    logical_cores = "N/A"

    if platform.system() == "Windows": # å¦‚æœæ˜¯ Windows ç³»ç»Ÿ
        cpu_info = platform.processor() # ä½¿ç”¨ platform.processor() è·å– CPU ä¿¡æ¯
        try:
            # è·å– Windows ä¸Šçš„æ ¸å¿ƒæ•°é‡ (éœ€è¦ WMI)
            import wmi
            c = wmi.WMI()
            for proc in c.Win32_Processor():
                physical_cores = proc.NumberOfCores
                logical_cores = proc.NumberOfLogicalProcessors
        except:
            pass # å¦‚æœ WMI ä¸å¯ç”¨ï¼Œå¿½ç•¥é”™è¯¯

    elif platform.system() == "Darwin": # å¦‚æœæ˜¯ macOS ç³»ç»Ÿ
        # åœ¨ macOS ä¸Šä½¿ç”¨ sysctl å‘½ä»¤è·å– CPU ä¿¡æ¯å’Œæ ¸å¿ƒæ•°é‡
        os.environ['PATH'] = os.environ['PATH'] + os.pathsep + '/usr/sbin' # æ›´æ–° PATH ç¯å¢ƒå˜é‡
        try:
            process_brand = subprocess.Popen(['sysctl', "machdep.cpu.brand_string"], stdout=subprocess.PIPE, stderr=subprocess.PIPE)
            stdout_brand, stderr_brand = process_brand.communicate()
            cpu_info = stdout_brand.decode().split(': ')[1].strip() if stdout_brand else "Could not retrieve CPU info"

            process_physical = subprocess.Popen(['sysctl', "hw.physicalcpu"], stdout=subprocess.PIPE, stderr=subprocess.PIPE)
            stdout_physical, stderr_physical = process_physical.communicate()
            physical_cores = stdout_physical.decode().split(': ')[1].strip() if stdout_physical else "N/A"

            process_logical = subprocess.Popen(['sysctl', "hw.logicalcpu"], stdout=subprocess.PIPE, stderr=subprocess.PIPE)
            stdout_logical, stderr_logical = process_logical.communicate()
            logical_cores = stdout_logical.decode().split(': ')[1].strip() if stdout_logical else "N/A"

        except:
            cpu_info = "Could not retrieve CPU info"
            physical_cores = "N/A"
            logical_cores = "N/A"

    else:  # Linux ç³»ç»Ÿ
        try:
            # åœ¨ Linux ä¸Šè¯»å– /proc/cpuinfo æ–‡ä»¶è·å– CPU ä¿¡æ¯å’Œæ ¸å¿ƒæ•°é‡
            with open('/proc/cpuinfo') as f:
                physical_cores_count = 0
                logical_cores_count = 0
                cpu_info_lines = []
                for line in f:
                    if line.startswith('model name'): # æŸ¥æ‰¾ä»¥ 'model name'å¼€å¤´çš„è¡Œ
                        if not cpu_info: # åªè·å–ç¬¬ä¸€ä¸ª model name
                            cpu_info = line.split(': ')[1].strip()
                    elif line.startswith('cpu cores'): # æŸ¥æ‰¾ä»¥ 'cpu cores' å¼€å¤´çš„è¡Œ
                        physical_cores_count = int(line.split(': ')[1].strip())
                    elif line.startswith('processor'): # æŸ¥æ‰¾ä»¥ 'processor' å¼€å¤´çš„è¡Œ
                        logical_cores_count += 1
                physical_cores = str(physical_cores_count) if physical_cores_count > 0 else "N/A"
                logical_cores = str(logical_cores_count) if logical_cores_count > 0 else "N/A"
                if not cpu_info:
                     cpu_info = "Could not retrieve CPU info"

        except:
            cpu_info = "Could not retrieve CPU info"
            physical_cores = "N/A"
            logical_cores = "N/A"

    return f"{cpu_info} ({physical_cores} physical cores, {logical_cores} logical cores)" # è¿”å› CPU ä¿¡æ¯å’Œæ ¸å¿ƒæ•°é‡


# è·å–å†…å­˜ä¿¡æ¯çš„å‡½æ•°
def get_memory_info():
    mem_info = "" # åˆå§‹åŒ–å†…å­˜ä¿¡æ¯å­—ç¬¦ä¸²
    if platform.system() == "Windows":
        # åœ¨ Windows ä¸Šä¸å®¹æ˜“é€šè¿‡æ ‡å‡†åº“è·å–ï¼Œéœ€è¦å¤–éƒ¨åº“æˆ– PowerShell
        mem_info = "Requires external tools on Windows" # è®¾ç½®æç¤ºä¿¡æ¯
    elif platform.system() == "Darwin": # å¦‚æœæ˜¯ macOS ç³»ç»Ÿ
        # åœ¨ macOS ä¸Šä½¿ç”¨ sysctl å‘½ä»¤è·å–å†…å­˜å¤§å°
        process = subprocess.Popen(['sysctl', "hw.memsize"], stdout=subprocess.PIPE, stderr=subprocess.PIPE) # è¿è¡Œ sysctl å‘½ä»¤
        stdout, stderr = process.communicate() # è·å–æ ‡å‡†è¾“å‡ºå’Œæ ‡å‡†é”™è¯¯
        mem_bytes = int(stdout.decode().split(': ')[1].strip()) # è§£æè¾“å‡ºï¼Œè·å–å†…å­˜å¤§å°ï¼ˆå­—èŠ‚ï¼‰
        mem_gb = mem_bytes / (1024**3) # è½¬æ¢ä¸º GB
        mem_info = f"{mem_gb:.2f} GB" # æ ¼å¼åŒ–è¾“å‡º
    else:  # Linux ç³»ç»Ÿ
        try:
            # åœ¨ Linux ä¸Šè¯»å– /proc/meminfo æ–‡ä»¶è·å–å†…å­˜ä¿¡æ¯
            with open('/proc/meminfo') as f:
                total_mem_kb = 0
                available_mem_kb = 0
                for line in f:
                    if line.startswith('MemTotal'): # æŸ¥æ‰¾ä»¥ 'MemTotal' å¼€å¤´çš„è¡Œ
                        total_mem_kb = int(line.split(':')[1].strip().split()[0]) # è§£æè¡Œï¼Œè·å–æ€»å†…å­˜ï¼ˆKBï¼‰
                    elif line.startswith('MemAvailable'): # æŸ¥æ‰¾ä»¥ 'MemAvailable' å¼€å¤´çš„è¡Œ
                         available_mem_kb = int(line.split(':')[1].strip().split()[0]) # è§£æè¡Œï¼Œè·å–å¯ç”¨å†…å­˜ï¼ˆKBï¼‰

                if total_mem_kb > 0:
                    total_mem_gb = total_mem_kb / (1024**2) # è½¬æ¢ä¸º GB
                    mem_info = f"{total_mem_gb:.2f} GB" # æ ¼å¼åŒ–è¾“å‡ºæ€»å†…å­˜
                    if available_mem_kb > 0:
                        available_mem_gb = available_mem_kb / (1024**2)
                        mem_info += f" (Available: {available_mem_gb:.2f} GB)" # æ·»åŠ å¯ç”¨å†…å­˜ä¿¡æ¯
                else:
                     mem_info = "Could not retrieve memory info" # å¦‚æœè¯»å–æ–‡ä»¶å‡ºé”™ï¼Œè®¾ç½®é”™è¯¯ä¿¡æ¯

        except:
            mem_info = "Could not retrieve memory info" # å¦‚æœè¯»å–æ–‡ä»¶å‡ºé”™ï¼Œè®¾ç½®é”™è¯¯ä¿¡æ¯
    return mem_info # è¿”å›å†…å­˜ä¿¡æ¯

# è·å– GPU ä¿¡æ¯çš„å‡½æ•°ï¼ŒåŒ…æ‹¬æ˜¾å­˜
def get_gpu_info():
    try:
        # å°è¯•ä½¿ç”¨ nvidia-smi è·å– NVIDIA GPU ä¿¡æ¯å’Œæ˜¾å­˜
        result = subprocess.run(['nvidia-smi', '--query-gpu=name,memory.total', '--format=csv,noheader'], capture_output=True, text=True)
        if result.returncode == 0: # å¦‚æœå‘½ä»¤æˆåŠŸæ‰§è¡Œ
            gpu_lines = result.stdout.strip().split('\n') # è§£æè¾“å‡ºï¼Œè·å– GPU åç§°å’Œæ˜¾å­˜
            gpu_info_list = []
            for line in gpu_lines:
                name, memory = line.split(', ')
                gpu_info_list.append(f"{name} ({memory})") # æ ¼å¼åŒ– GPU ä¿¡æ¯
            return ", ".join(gpu_info_list) if gpu_info_list else "NVIDIA GPU found, but info not listed" # è¿”å› GPU ä¿¡æ¯æˆ–æç¤ºä¿¡æ¯
        else:
             # å°è¯•ä½¿ç”¨ lshw è·å–å…¶ä»– GPU ä¿¡æ¯ (éœ€è¦å®‰è£… lshw)
            try:
                result_lshw = subprocess.run(['lshw', '-C', 'display'], capture_output=True, text=True)
                if result_lshw.returncode == 0: # å¦‚æœå‘½ä»¤æˆåŠŸæ‰§è¡Œ
                     # ç®€å•è§£æè¾“å‡ºä¸­çš„ product åç§°å’Œæ˜¾å­˜
                    gpu_info_lines = []
                    current_gpu = {}
                    for line in result_lshw.stdout.splitlines():
                        if 'product:' in line:
                             if current_gpu:
                                 gpu_info_lines.append(f"{current_gpu.get('product', 'GPU')} ({current_gpu.get('memory', 'N/A')})")
                             current_gpu = {'product': line.split('product:')[1].strip()}
                        elif 'size:' in line and 'memory' in line:
                             current_gpu['memory'] = line.split('size:')[1].strip()

                    if current_gpu: # æ·»åŠ æœ€åä¸€ä¸ª GPU çš„ä¿¡æ¯
                        gpu_info_lines.append(f"{current_gpu.get('product', 'GPU')} ({current_gpu.get('memory', 'N/A')})")

                    return ", ".join(gpu_info_lines) if gpu_info_lines else "GPU found (via lshw), but info not parsed" # å¦‚æœæ‰¾åˆ° GPU ä½†ä¿¡æ¯æ— æ³•è§£æï¼Œè®¾ç½®æç¤ºä¿¡æ¯
                else:
                    return "No GPU found (checked nvidia-smi and lshw)" # å¦‚æœä¸¤ä¸ªå‘½ä»¤éƒ½æ‰¾ä¸åˆ° GPUï¼Œè®¾ç½®æç¤ºä¿¡æ¯
            except FileNotFoundError:
                 return "No GPU found (checked nvidia-smi, lshw not found)" # å¦‚æœæ‰¾ä¸åˆ° lshw å‘½ä»¤ï¼Œè®¾ç½®æç¤ºä¿¡æ¯
    except FileNotFoundError:
        return "No GPU found (nvidia-smi not found)" # å¦‚æœæ‰¾ä¸åˆ° nvidia-smi å‘½ä»¤ï¼Œè®¾ç½®æç¤ºä¿¡æ¯


# è·å– CUDA ç‰ˆæœ¬çš„å‡½æ•°
def get_cuda_version():
    try:
        # å°è¯•ä½¿ç”¨ nvcc --version è·å– CUDA ç‰ˆæœ¬
        result = subprocess.run(['nvcc', '--version'], capture_output=True, text=True)
        if result.returncode == 0: # å¦‚æœå‘½ä»¤æˆåŠŸæ‰§è¡Œ
            for line in result.stdout.splitlines():
                if 'release' in line: # æŸ¥æ‰¾åŒ…å« 'release' çš„è¡Œ
                    return line.split('release ')[1].split(',')[0] # è§£æè¡Œï¼Œæå–ç‰ˆæœ¬å·
        return "CUDA not found or version not parsed" # å¦‚æœæ‰¾ä¸åˆ° CUDA æˆ–ç‰ˆæœ¬æ— æ³•è§£æï¼Œè®¾ç½®æç¤ºä¿¡æ¯
    except FileNotFoundError:
        return "CUDA not found" # å¦‚æœæ‰¾ä¸åˆ° nvcc å‘½ä»¤ï¼Œè®¾ç½®æç¤ºä¿¡æ¯

# è·å– Python ç‰ˆæœ¬çš„å‡½æ•°
def get_python_version():
    return platform.python_version() # è·å– Python ç‰ˆæœ¬

# è·å– Conda ç‰ˆæœ¬çš„å‡½æ•°
def get_conda_version():
    try:
        # å°è¯•ä½¿ç”¨ conda --version è·å– Conda ç‰ˆæœ¬
        result = subprocess.run(['conda', '--version'], capture_output=True, text=True)
        if result.returncode == 0: # å¦‚æœå‘½ä»¤æˆåŠŸæ‰§è¡Œ
            return result.stdout.strip() # è¿”å› Conda ç‰ˆæœ¬
        return "Conda not found or version not parsed" # å¦‚æœæ‰¾ä¸åˆ° Conda æˆ–ç‰ˆæœ¬æ— æ³•è§£æï¼Œè®¾ç½®æç¤ºä¿¡æ¯
    except FileNotFoundError:
        return "Conda not found" # å¦‚æœæ‰¾ä¸åˆ° conda å‘½ä»¤ï¼Œè®¾ç½®æç¤ºä¿¡æ¯

# è·å–ç‰©ç†ç£ç›˜ç©ºé—´ä¿¡æ¯çš„å‡½æ•°
def get_disk_space():
    try:
        total, used, free = shutil.disk_usage("/") # è·å–æ ¹ç›®å½•çš„ç£ç›˜ä½¿ç”¨æƒ…å†µ
        total_gb = total / (1024**3) # è½¬æ¢ä¸º GB
        used_gb = used / (1024**3) # è½¬æ¢ä¸º GB
        free_gb = free / (1024**3) # è½¬æ¢ä¸º GB
        return f"Total: {total_gb:.2f} GB, Used: {used_gb:.2f} GB, Free: {free_gb:.2f} GB" # æ ¼å¼åŒ–è¾“å‡º
    except Exception as e:
        return f"Could not retrieve disk info: {e}" # å¦‚æœè·å–ä¿¡æ¯å‡ºé”™ï¼Œè®¾ç½®é”™è¯¯ä¿¡æ¯

# è·å–ç¯å¢ƒä¿¡æ¯
os_name = platform.system() # è·å–æ“ä½œç³»ç»Ÿåç§°
os_version = platform.release() # è·å–æ“ä½œç³»ç»Ÿç‰ˆæœ¬
if os_name == "Linux":
    try:
        # åœ¨ Linux ä¸Šå°è¯•è·å–å‘è¡Œç‰ˆå’Œç‰ˆæœ¬
        lsb_info = subprocess.run(['lsb_release', '-a'], capture_output=True, text=True)
        if lsb_info.returncode == 0: # å¦‚æœå‘½ä»¤æˆåŠŸæ‰§è¡Œ
            for line in lsb_info.stdout.splitlines():
                if 'Description:' in line: # æŸ¥æ‰¾åŒ…å« 'Description:' çš„è¡Œ
                    os_version = line.split('Description:')[1].strip() # æå–æè¿°ä¿¡æ¯ä½œä¸ºç‰ˆæœ¬
                    break # æ‰¾åˆ°åé€€å‡ºå¾ªç¯
                elif 'Release:' in line: # æŸ¥æ‰¾åŒ…å« 'Release:' çš„è¡Œ
                     os_version = line.split('Release:')[1].strip() # æå–ç‰ˆæœ¬å·
                     # å°è¯•è·å– codename
                     try:
                         codename_info = subprocess.run(['lsb_release', '-c'], capture_output=True, text=True)
                         if codename_info.returncode == 0:
                             os_version += f" ({codename_info.stdout.split(':')[1].strip()})" # å°† codename æ·»åŠ åˆ°ç‰ˆæœ¬ä¿¡æ¯ä¸­
                     except:
                         pass # å¦‚æœè·å– codename å¤±è´¥åˆ™å¿½ç•¥

    except FileNotFoundError:
        pass # lsb_release å¯èƒ½æœªå®‰è£…ï¼Œå¿½ç•¥é”™è¯¯

full_os_info = f"{os_name} {os_version}" # ç»„åˆå®Œæ•´çš„æ“ä½œç³»ç»Ÿä¿¡æ¯
cpu_info = get_cpu_info() # è°ƒç”¨å‡½æ•°è·å– CPU ä¿¡æ¯å’Œæ ¸å¿ƒæ•°é‡
memory_info = get_memory_info() # è°ƒç”¨å‡½æ•°è·å–å†…å­˜ä¿¡æ¯
gpu_info = get_gpu_info() # è°ƒç”¨å‡½æ•°è·å– GPU ä¿¡æ¯å’Œæ˜¾å­˜
cuda_version = get_cuda_version() # è°ƒç”¨å‡½æ•°è·å– CUDA ç‰ˆæœ¬
python_version = get_python_version() # è°ƒç”¨å‡½æ•°è·å– Python ç‰ˆæœ¬
conda_version = get_conda_version() # è°ƒç”¨å‡½æ•°è·å– Conda ç‰ˆæœ¬
disk_info = get_disk_space() # è°ƒç”¨å‡½æ•°è·å–ç‰©ç†ç£ç›˜ç©ºé—´ä¿¡æ¯


# åˆ›å»ºç”¨äºå­˜å‚¨æ•°æ®çš„å­—å…¸
env_data = {
    "é¡¹ç›®": [ # é¡¹ç›®åç§°åˆ—è¡¨
        "æ“ä½œç³»ç»Ÿ",
        "CPU ä¿¡æ¯",
        "å†…å­˜ä¿¡æ¯",
        "GPU ä¿¡æ¯",
        "CUDA ä¿¡æ¯",
        "Python ç‰ˆæœ¬",
        "Conda ç‰ˆæœ¬",
        "ç‰©ç†ç£ç›˜ç©ºé—´" # æ·»åŠ ç‰©ç†ç£ç›˜ç©ºé—´
    ],
    "ä¿¡æ¯": [ # å¯¹åº”çš„ä¿¡æ¯åˆ—è¡¨
        full_os_info,
        cpu_info,
        memory_info,
        gpu_info,
        cuda_version,
        python_version,
        conda_version,
        disk_info # æ·»åŠ ç‰©ç†ç£ç›˜ç©ºé—´ä¿¡æ¯
    ]
}

# åˆ›å»ºä¸€ä¸ª pandas DataFrame
df = pd.DataFrame(env_data)

# æ‰“å°è¡¨æ ¼
print("### ç¯å¢ƒä¿¡æ¯") # æ‰“å°æ ‡é¢˜
print(df.to_markdown(index=False)) # å°† DataFrame è½¬æ¢ä¸º Markdown æ ¼å¼å¹¶æ‰“å°ï¼Œä¸åŒ…å«ç´¢å¼•
```

```python
# å¤§æ¨¡å‹è®¾ç½®
import os
import time
from typing import Dict, List, Tuple, Any, Optional

# å¯¼å…¥ matplotlib ç”¨äºæ•°æ®å¯è§†åŒ–ï¼Œå¦‚æœç¯å¢ƒä¸­æ²¡æœ‰åˆ™è·³è¿‡
try:
    import matplotlib.pyplot as plt
    HAS_MPL = True
except Exception:
    plt = None
    HAS_MPL = False

# OpenAI API ç¯å¢ƒé…ç½®
# è®¾ç½®é»˜è®¤ä½¿ç”¨çš„ OpenAI æ¨¡å‹
DEFAULT_MODEL = os.getenv("OPENAI_MODEL", "gpt-4o-mini")
# è·å– OpenAI API Key
OPENAI_API_KEY = os.getenv("OPENAI_API_KEY","sk-R2utGqjya03Bgnc77273729e7b73457e97A140D79bF3B944") # Added API Key variable
# è®¾ç½® OpenAI API çš„åŸºç¡€ URL
OPENAI_BASE_URL = os.getenv("OPENAI_BASE_URL","https://api.apiyi.com/v1") # Added Base URL variable


from openai import OpenAI
# åˆå§‹åŒ– OpenAI å®¢æˆ·ç«¯ï¼Œä¼ å…¥ API Key å’ŒåŸºç¡€ URL
client: OpenAI = OpenAI(api_key=OPENAI_API_KEY, base_url=OPENAI_BASE_URL) # Added api_key and base_url


# tiktoken å¯é€‰åº“ï¼Œç”¨äºæ›´ç²¾ç¡®åœ°è®¡ç®— Token æ•°é‡
# å¦‚æœ tiktoken æ²¡æœ‰å®‰è£…ï¼Œåˆ™é€€åŒ–ä¸ºç²—ç•¥çš„ Token ä¼°è®¡æ–¹æ³•
try:
    import tiktoken
    # æ ¹æ®æ¨¡å‹è·å– tiktoken ç¼–ç å™¨
    tokenizer = tiktoken.encoding_for_model(DEFAULT_MODEL)
except Exception:
    # å¦‚æœ tiktoken å¯¼å…¥å¤±è´¥ï¼Œè®¾ç½® tokenizer ä¸º None
    tokenizer = None

# è®¡ç®—ç»™å®šæ–‡æœ¬çš„ Token æ•°é‡
def count_tokens(text: str) -> int:
    if tokenizer is not None:
        # å¦‚æœ tiktoken å¯ç”¨ï¼Œä½¿ç”¨ tiktoken ç²¾ç¡®è®¡ç®—
        return len(tokenizer.encode(text))
    # å¦‚æœ tiktoken ä¸å¯ç”¨ï¼Œä½¿ç”¨ç²—ç•¥çš„æ–¹æ³•ä¼°è®¡ Token æ•°é‡ï¼ˆæŒ‰è¯æ•° * 1.3ï¼‰
    return int(len(text.split()) * 1.3)

# æµ‹é‡å‡½æ•°çš„æ‰§è¡Œæ—¶é—´
# def measure_latency(func, *args, **kwargs) -> Tuple[Any, float]:
#     t0 = time.time() # è®°å½•å¼€å§‹æ—¶é—´
#     res = func(*args, **kwargs) # æ‰§è¡Œå‡½æ•°
#     t1 = time.time() # è®°å½•ç»“æŸæ—¶é—´
#     return res, (t1 - t0) # è¿”å›å‡½æ•°ç»“æœå’Œæ‰§è¡Œæ—¶é—´
```

```python
# è°ƒç”¨å¤§æ¨¡å‹ç”Ÿæˆå“åº”
def generate_response(prompt: str) -> Tuple[str, float]:
    # å®¢æˆ·ç«¯å·²åˆå§‹åŒ–
    try:
        t0 = time.time() # è®°å½•è°ƒç”¨å¼€å§‹æ—¶é—´
        # è°ƒç”¨ OpenAI Chat Completion API
        resp = client.chat.completions.create(
            model=DEFAULT_MODEL, # æŒ‡å®šä½¿ç”¨çš„æ¨¡å‹
            messages=[{"role": "user", "content": prompt}], # ä¼ é€’ç”¨æˆ·æç¤º
            temperature=0.7, # è®¾ç½®ç”Ÿæˆçš„å¤šæ ·æ€§ï¼ˆ0.0-1.0ï¼Œè¶Šé«˜è¶Šéšæœºï¼‰
            max_tokens=800, # è®¾ç½®å“åº”çš„æœ€å¤§ Token æ•°é‡
        )
        latency = time.time() - t0 # è®¡ç®—è°ƒç”¨æ—¶å»¶
        # è¿”å›ç”Ÿæˆçš„æ–‡æœ¬å†…å®¹å’Œæ—¶å»¶
        return resp.choices[0].message.content or "", latency
    except Exception as e:
        # å¦‚æœè°ƒç”¨å¤±è´¥ï¼Œè¿”å›é”™è¯¯ä¿¡æ¯å’Œ 0 æ—¶å»¶
        return f"[OpenAI è°ƒç”¨å¤±è´¥ï¼š{e}]", 0.0

# è®¡ç®—ä¸æ¨¡å‹äº¤äº’ç›¸å…³çš„æŒ‡æ ‡
def calculate_metrics(prompt: str, response: str, latency: float) -> Dict[str, float]:
    pt = count_tokens(prompt) # è®¡ç®—æç¤ºçš„ Token æ•°é‡
    rt = count_tokens(response) # è®¡ç®—å“åº”çš„ Token æ•°é‡

    # è®¡ç®— Token æ•ˆç‡ï¼ˆå“åº” Token / æç¤º Tokenï¼‰ï¼Œæç¤º Token ä¸º 0 æ—¶æ•ˆç‡ä¸º 0
    eff = (rt / pt) if pt > 0 else 0.0

    # è®¡ç®—æ¯åƒ Token çš„æ—¶å»¶ï¼ˆæ—¶å»¶ / æç¤º Token * 1000ï¼‰ï¼Œæç¤º Token ä¸º 0 æ—¶æ—¶å»¶ä¸º 0
    latency_per_1k = (latency / pt) * 1000 if pt > 0 else 0.0

    # è¿”å›åŒ…å«å„ç§æŒ‡æ ‡çš„å­—å…¸
    return {
        "prompt_tokens": pt,
        "response_tokens": rt,
        "token_efficiency": eff,
        "latency": latency,
        # å°†æ¯åƒ Token çš„æ—¶å»¶ï¼ˆç§’ï¼‰ï¼Œå¹¶å››èˆäº”å…¥åˆ°ä¸¤ä½å°æ•°
        "latency_per_1k": round(latency_per_1k, 2), # Convert to ms and round
    }
```

```python
# å®šä¹‰ä¸€ä¸ªåŸºç¡€æç¤º
base_prompt = "è¯·å†™ä¸€æ®µå…³äºæ—…æ¸¸è·¯çº¿è§„åˆ’çš„è¯´æ˜æ–‡å­—ã€‚"

# åˆ›å»ºä¸åŒç±»å‹çš„æ‰©å±•æç¤ºï¼Œé€šè¿‡æ·»åŠ ä¸åŒçš„ä¸Šä¸‹æ–‡ä¿¡æ¯
expanded_prompts = {
    # åŸºç¡€æç¤ºï¼Œä¸åŒ…å«ä»»ä½•é¢å¤–ä¸Šä¸‹æ–‡
    "base": base_prompt,
    # æ·»åŠ â€œè§’è‰²â€ä¸Šä¸‹æ–‡ï¼šæŒ‡å®šæ¨¡å‹æ‰®æ¼”çš„è§’è‰²
    "with_role": "ä½ æ˜¯ä¸€åä¸“ä¸šçš„æ—…æ¸¸è§„åˆ’å¸ˆã€‚\nè¯·å†™ä¸€æ®µå…³äºæ—…æ¸¸è·¯çº¿è§„åˆ’çš„è¯´æ˜æ–‡å­—ã€‚",
    # æ·»åŠ â€œç¤ºä¾‹â€ä¸Šä¸‹æ–‡ï¼šæä¾›æœŸæœ›è¾“å‡ºçš„ä¾‹å­ï¼ˆFew-shot learningï¼‰
    "with_examples": (
        "è¯·å†™ä¸€æ®µå…³äºæ—…æ¸¸è·¯çº¿è§„åˆ’çš„è¯´æ˜æ–‡å­—ã€‚\n\n"
        "ç¤ºä¾‹ 1ï¼š\næ—…æ¸¸è·¯çº¿è§„åˆ’æ˜¯åˆ¶å®šæ—…è¡Œè¡Œç¨‹çš„è¿‡ç¨‹ï¼ŒåŒ…æ‹¬é€‰æ‹©ç›®çš„åœ°ã€äº¤é€šæ–¹å¼ã€ä½å®¿ã€æ™¯ç‚¹å’Œæ´»åŠ¨ç­‰ã€‚ä¸€ä¸ªå¥½çš„è§„åˆ’èƒ½å¸®åŠ©æ—…è¡Œè€…å……åˆ†åˆ©ç”¨æ—¶é—´ï¼Œæ§åˆ¶é¢„ç®—ï¼Œå¹¶è·å¾—æ›´å¥½çš„æ—…è¡Œä½“éªŒã€‚\n\n"
        "ç¤ºä¾‹ 2ï¼š\nè§„åˆ’æ—…æ¸¸è·¯çº¿æ—¶ï¼Œéœ€è¦è€ƒè™‘æ—…è¡Œæ—¶é—´ã€é¢„ç®—ã€æ—…è¡Œåå¥½ï¼ˆå¦‚æ–‡åŒ–ã€è‡ªç„¶ã€æ¢é™©ï¼‰ã€åŒè¡Œäººå‘˜ä»¥åŠå½“åœ°çš„æ°”å€™å’Œå­£èŠ‚ã€‚è¯¦ç»†çš„è§„åˆ’å¯ä»¥åŒ…æ‹¬æ¯æ—¥è¡Œç¨‹ã€é¢„è®¢ä¿¡æ¯å’Œå¤‡ç”¨æ–¹æ¡ˆã€‚"
    ),
    # æ·»åŠ â€œçº¦æŸâ€ä¸Šä¸‹æ–‡ï¼šæŒ‡å®šè¾“å‡ºéœ€è¦æ»¡è¶³çš„æ¡ä»¶æˆ–è§„åˆ™
    "with_constraints": (
        "è¯·å†™ä¸€æ®µå…³äºæ—…æ¸¸è·¯çº¿è§„åˆ’çš„è¯´æ˜æ–‡å­—ã€‚\n"
        "- è‡³å°‘åŒ…å«ä¸‰ä¸ªå…³é”®è€ƒè™‘å› ç´ ï¼ˆå¦‚é¢„ç®—ã€æ—¶é—´ã€åå¥½ï¼‰\n"
        "- æåŠè§„åˆ’çš„å¥½å¤„\n"
        "- ä»¥â€˜å¼€å§‹ä½ çš„è§„åˆ’å§ï¼â€™ç»“å°¾\n"
        "- ä¿æŒå®ç”¨ä¸”æ˜“äºç†è§£çš„è¯­æ°”"
    ),
    # æ·»åŠ â€œå—ä¼—â€ä¸Šä¸‹æ–‡ï¼šæŒ‡å®šç›®æ ‡è¯»è€…æˆ–ç”¨æˆ·
    "with_audience": (
        "è¯·ä¸ºç¬¬ä¸€æ¬¡è®¡åˆ’è‡ªç”±è¡Œçš„å¤§å­¦ç”Ÿï¼Œå†™ä¸€æ®µå…³äºæ—…æ¸¸è·¯çº¿è§„åˆ’çš„è¯´æ˜æ–‡å­—ï¼Œ\n"
        "å°½é‡ä½¿ç”¨ç®€å•æ˜“æ‡‚çš„è¯­è¨€å’Œå®ç”¨çš„å»ºè®®ã€‚"
    ),
    # ç»¼åˆæç¤ºï¼šç»“åˆäº†å¤šç§ä¸Šä¸‹æ–‡å±‚ï¼Œæ„å»ºä¸€ä¸ªä¿¡æ¯æ›´å…¨é¢çš„æç¤º
    "comprehensive": (
        "ä½ æ˜¯ä¸€åä¸“ä¸šçš„æ—…æ¸¸è§„åˆ’å¸ˆã€‚\n\n" # è§’è‰²
        "è¯·ä¸ºç¬¬ä¸€æ¬¡è®¡åˆ’è‡ªç”±è¡Œçš„å¤§å­¦ç”Ÿï¼Œå†™ä¸€æ®µå…³äºæ—…æ¸¸è·¯çº¿è§„åˆ’çš„è¯´æ˜æ–‡å­—ï¼Œ\n" # åŸºç¡€ä»»åŠ¡ä¸å—ä¼—
        "ä½¿ç”¨ç®€å•æ˜“æ‡‚çš„è¯­è¨€å’Œå®ç”¨çš„å»ºè®®ã€‚\n\n" # å—ä¼—ç›¸å…³è¦æ±‚
        "å†™ä½œæŒ‡å—ï¼š\n" # çº¦æŸ/æŒ‡å—
        "- è‡³å°‘åŒ…å«ä¸‰ä¸ªå…³é”®è€ƒè™‘å› ç´ ï¼ˆå¦‚é¢„ç®—ã€æ—¶é—´ã€åå¥½ï¼‰\n"
        "- æåŠè§„åˆ’çš„å¥½å¤„\n"
        "- ä»¥â€˜å¼€å§‹ä½ çš„è§„åˆ’å§ï¼â€™ç»“å°¾\n"
        "- ä¿æŒå®ç”¨ä¸”æ˜“äºç†è§£çš„è¯­æ°”\n\n"
        "è¯­æ°”ä¸ç»“æ„ç¤ºä¾‹ï¼š\n" # ç¤ºä¾‹ï¼ˆæä¾›è¾“å‡ºçš„é£æ ¼å’Œç»“æ„å‚è€ƒï¼‰
        "è®¡åˆ’ä¸€æ¬¡æ—…è¡Œå°±åƒæ˜¯ç”»ä¸€å¼ å¯»å®å›¾ï¼Œæ¯ä¸€æ­¥éƒ½å¾ˆé‡è¦ã€‚é¦–å…ˆï¼Œå®šå¥½ä½ çš„ç›®çš„åœ°å’Œå¤§æ¦‚çš„æ—…è¡Œæ—¥æœŸã€‚ç„¶åï¼Œè€ƒè™‘ä½ çš„é¢„ç®—ï¼Œè¿™ä¼šå½±å“ä½ çš„äº¤é€šå’Œä½å®¿é€‰æ‹©ã€‚åˆ«å¿˜äº†ä½ çš„å…´è¶£ï¼Œæ˜¯å–œæ¬¢è‡ªç„¶é£å…‰è¿˜æ˜¯åŸå¸‚æ¢ç´¢ï¼Ÿ"
    ),
}
```

```python
# ç”¨äºå­˜å‚¨ä¸åŒæç¤ºçš„æµ‹è¯•ç»“æœ
results = {}
# ç”¨äºå­˜å‚¨ä¸åŒæç¤ºç”Ÿæˆçš„å“åº”
responses = {}

# éå†æ¯ç§æ‰©å±•æç¤ºè¿›è¡Œæµ‹è¯•
for name, p in expanded_prompts.items():
    print(f"æµ‹è¯•ï¼š{name}")
    # è°ƒç”¨ç”Ÿæˆå“åº”å‡½æ•°ï¼Œå¹¶æµ‹é‡æ—¶å»¶
    resp, lat = generate_response(p)
    # å­˜å‚¨ç”Ÿæˆçš„å“åº”
    responses[name] = resp
    # è®¡ç®—å¹¶å­˜å‚¨å„ç§æŒ‡æ ‡ï¼ˆToken æ•°ã€æ•ˆç‡ã€æ—¶å»¶ç­‰ï¼‰
    results[name] = calculate_metrics(p, resp, lat)
    # æ‰“å°æµ‹è¯•ç»“æœæ‘˜è¦
    print("  æç¤º tokenï¼š", results[name]["prompt_tokens"])
    print("  å“åº” tokenï¼š", results[name]["response_tokens"])
    print("  æ—¶å»¶ï¼š{:.2f}s".format(results[name]["latency"]))
    print("-" * 40)
```

### 2. å¯è§†åŒ–ä¸åˆ†æï¼ˆToken/æ•ˆç‡/æ—¶å»¶ï¼‰

è¿è¡Œå•å…ƒæ ¼åï¼Œæ‚¨ä¼šçœ‹åˆ°ä¸€ä¸ªåŒ…å«å››ä¸ªå­å›¾çš„å›¾è¡¨ã€‚è¿™ä¸ªå›¾è¡¨å°†æˆ‘ä»¬ä¹‹å‰æµ‹è¯•ä¸åŒæç¤ºï¼ˆPromptï¼‰å¾—åˆ°çš„ç»“æœç”¨å›¾è¡¨çš„æ–¹å¼å±•ç¤ºå‡ºæ¥ï¼Œå¸®åŠ©æˆ‘ä»¬æ›´ç›´è§‚åœ°æ¯”è¾ƒä¸åŒä¸Šä¸‹æ–‡æ‰©å±•ç­–ç•¥çš„æ•ˆæœã€‚

å¯¹äºå¤§æ¨¡å‹æŠ€æœ¯çš„åˆå­¦è€…æ¥è¯´ï¼Œç†è§£è¿™äº›æŒ‡æ ‡éå¸¸é‡è¦ï¼Œå› ä¸ºå®ƒä»¬ç›´æ¥å…³ç³»åˆ°ä½¿ç”¨å¤§æ¨¡å‹çš„**æˆæœ¬**ï¼ˆToken æ•°é‡ï¼‰å’Œ**æ•ˆç‡**ï¼ˆå“åº”é€Ÿåº¦å’Œç”Ÿæˆå†…å®¹çš„æ¯”ä¾‹ï¼‰ã€‚

```python
# ä» results å­—å…¸ä¸­æå–ç”¨äºç»˜å›¾çš„æ•°æ®
prompt_types = list(results.keys()) # æç¤ºç±»å‹åç§°åˆ—è¡¨
prompt_tokens = [results[k]['prompt_tokens'] for k in prompt_types] # å„ç§æç¤ºçš„ Token æ•°é‡
response_tokens = [results[k]['response_tokens'] for k in prompt_types] # å„ç§å“åº”çš„ Token æ•°é‡
latencies = [results[k]['latency'] for k in prompt_types] # å„ç§å“åº”çš„æ—¶å»¶

# å®šä¹‰ä¸€ä¸ªå‡½æ•°ç”¨äºç»˜åˆ¶æŒ‡æ ‡çš„æŸ±çŠ¶å›¾
def plot_bars():
    # åˆ›å»ºä¸€ä¸ªåŒ…å« 2x2 å­å›¾çš„å›¾è¡¨
    fig, axes = plt.subplots(2, 2, figsize=(14, 10))

    # ç¬¬ä¸€ä¸ªå­å›¾ï¼šToken Usage (æç¤º Token + å“åº” Token)
    # Token Usage: æ˜¾ç¤ºä¸åŒæç¤ºç±»å‹ä¸‹ï¼Œå‘é€ç»™æ¨¡å‹çš„æç¤º Token æ•°é‡å’Œæ¨¡å‹è¿”å›çš„å“åº” Token æ•°é‡ã€‚
    # Token æ˜¯å¤§æ¨¡å‹å¤„ç†æ–‡æœ¬çš„åŸºæœ¬å•ä½ï¼ŒToken æ•°é‡ç›´æ¥å½±å“æˆæœ¬å’Œå¤„ç†æ—¶é—´ã€‚
    axes[0, 0].bar(prompt_types, prompt_tokens, label='Prompt Tokens', alpha=0.7, color='blue')
    axes[0, 0].bar(prompt_types, response_tokens, bottom=prompt_tokens, label='Response Tokens', alpha=0.7, color='green')
    axes[0, 0].set_title('Token Usage by Prompt Type')
    axes[0, 0].set_ylabel('Number of Tokens')
    axes[0, 0].legend()
    # æ—‹è½¬ x è½´æ ‡ç­¾ï¼Œé¿å…é‡å 
    plt.setp(axes[0, 0].get_xticklabels(), rotation=45, ha='right')

    # ç¬¬äºŒä¸ªå­å›¾ï¼šToken Efficiency (Response/Prompt)
    # Token Efficiency: å“åº” Token æ•°é‡é™¤ä»¥æç¤º Token æ•°é‡ã€‚
    # è¡¨ç¤ºæ¨¡å‹ç”Ÿæˆå“åº”çš„æ•ˆç‡ï¼Œé«˜æ•ˆç‡æ„å‘³ç€ç”¨è¾ƒå°‘çš„æç¤º Token è·å¾—äº†è¾ƒå¤šçš„å“åº” Tokenï¼Œè¿™é€šå¸¸æ˜¯æœŸæœ›çš„ã€‚
    token_efficiency = [results[k]['token_efficiency'] for k in prompt_types]
    axes[0, 1].bar(prompt_types, token_efficiency, color='purple', alpha=0.7)
    axes[0, 1].set_title('Token Efficiency (Response/Prompt)')
    axes[0, 1].set_ylabel('Efficiency Ratio')
    # æ—‹è½¬ x è½´æ ‡ç­¾
    plt.setp(axes[0, 1].get_xticklabels(), rotation=45, ha='right')

    # ç¬¬ä¸‰ä¸ªå­å›¾ï¼šResponse Latency
    # Response Latency: ä»å‘é€æç¤ºåˆ°æ¥æ”¶å®Œæ•´å“åº”æ‰€éœ€çš„æ—¶é—´ï¼ˆä»¥ç§’ä¸ºå•ä½ï¼‰ã€‚
    # åæ˜ æ¨¡å‹çš„å“åº”é€Ÿåº¦ï¼Œæ—¶å»¶è¶Šä½ï¼Œç”¨æˆ·ä½“éªŒè¶Šå¥½ã€‚
    latencies = [results[k]['latency'] for k in prompt_types]
    axes[1, 0].bar(prompt_types, latencies, color='red', alpha=0.7)
    axes[1, 0].set_title('Response Latency')
    axes[1, 0].set_ylabel('Seconds')
    # æ—‹è½¬ x è½´æ ‡ç­¾
    plt.setp(axes[1, 0].get_xticklabels(), rotation=45, ha='right')

    # ç¬¬å››ä¸ªå­å›¾ï¼šLatency per 1k Tokens
    # Latency per 1k Tokens: å°†æ€»æ—¶å»¶æ ‡å‡†åŒ–åˆ°æ¯å¤„ç† 1000 ä¸ªæç¤º Token æ‰€éœ€çš„æ—¶é—´ï¼ˆä»¥æ¯«ç§’ä¸ºå•ä½ï¼‰ã€‚
    # è¿™ä¸ªæŒ‡æ ‡æœ‰åŠ©äºæ¯”è¾ƒä¸åŒæç¤ºé•¿åº¦ä¸‹çš„æ¨¡å‹æ€§èƒ½ï¼Œæ’é™¤äº†å› æç¤ºé•¿åº¦ä¸åŒå¸¦æ¥çš„ç»å¯¹æ—¶å»¶å·®å¼‚ã€‚
    latency_per_1k = [results[k]['latency_per_1k'] for k in prompt_types]
    axes[1, 1].bar(prompt_types, latency_per_1k, color='orange', alpha=0.7)
    axes[1, 1].set_title('Latency per 1k Tokens')
    axes[1, 1].set_ylabel('Seconds per 1k Tokens')
    # æ—‹è½¬ x è½´æ ‡ç­¾
    plt.setp(axes[1, 1].get_xticklabels(), rotation=45, ha='right')

    # è‡ªåŠ¨è°ƒæ•´å­å›¾å¸ƒå±€ï¼Œé¿å…é‡å 
    plt.tight_layout()

# æ£€æŸ¥æ˜¯å¦å®‰è£…äº† matplotlibï¼Œå¦‚æœå®‰è£…äº†åˆ™ç»˜åˆ¶å›¾è¡¨
if HAS_MPL:
    plot_bars(); # è°ƒç”¨ç»˜å›¾å‡½æ•°
    plt.show() # æ˜¾ç¤ºå›¾è¡¨
else:
    print('[æœªå®‰è£… matplotlibï¼Œè·³è¿‡å¯è§†åŒ–]')
```

**è¾“å‡ºå›¾è¡¨è§£é‡Šï¼š**

è¾“å‡ºç»“æœæ˜¾ç¤ºäº†å››ä¸ªæŸ±çŠ¶å›¾ï¼Œæ¯ä¸ªå›¾éƒ½å±•ç¤ºäº†ä¸åŒâ€œPrompt Typeâ€ï¼ˆå°±æ˜¯æˆ‘ä»¬ä¹‹å‰å®šä¹‰çš„å„ç§æç¤ºï¼Œå¦‚ baseã€with_roleã€comprehensive ç­‰ï¼‰åœ¨æŸä¸ªæŒ‡æ ‡ä¸Šçš„è¡¨ç°ã€‚

2. Token Efficiency (Response/Prompt)
   - **å›¾è¡¨å†…å®¹ï¼š** è¿™ä¸ªå›¾æ˜¾ç¤ºäº†æ¨¡å‹å“åº”çš„ Token æ•°é‡é™¤ä»¥æç¤ºçš„ Token æ•°é‡çš„æ¯”å€¼ã€‚
   - **è§£é‡Šï¼š** è¿™ä¸ªæŒ‡æ ‡å¯ä»¥ç²—ç•¥è¡¡é‡æ¨¡å‹åœ¨ç»™å®šæç¤ºä¸‹ç”Ÿæˆâ€œæœ‰ç”¨â€å†…å®¹çš„æ•ˆç‡ã€‚æ¯”å€¼è¶Šé«˜ï¼Œæ„å‘³ç€æ¨¡å‹ç”¨ç›¸å¯¹è¾ƒå°‘çš„æç¤º Token ç”Ÿæˆäº†è¾ƒå¤šçš„å“åº” Tokenã€‚é€šå¸¸æˆ‘ä»¬å¸Œæœ›è¿™ä¸ªå€¼é«˜ä¸€äº›ï¼Œä½†å¹¶éè¶Šé«˜è¶Šå¥½ï¼Œå…³é”®æ˜¯ç”Ÿæˆçš„å“åº”è¦æœ‰**è´¨é‡**ä¸”**ç›¸å…³**ã€‚åŸºç¡€æç¤º (`base`) çš„æ•ˆç‡çœ‹èµ·æ¥å¾ˆé«˜ï¼Œä½†è¿™å¯èƒ½æ˜¯å› ä¸ºå®ƒæç¤ºéå¸¸ç®€çŸ­ï¼Œæ¨¡å‹è‡ªç”±å‘æŒ¥ç©ºé—´å¤§ï¼Œç”Ÿæˆçš„å“åº”ç›¸å¯¹è¾ƒé•¿ï¼›è€ŒåŒ…å«ç¤ºä¾‹ (`with_examples`) çš„æç¤ºï¼Œæç¤ºæœ¬èº«è¾ƒé•¿ï¼Œä½†å“åº”å¯èƒ½æ›´èšç„¦ï¼Œæ‰€ä»¥æ•ˆç‡æ¯” `base` ä½ã€‚
3. Response Latency (æ—¶å»¶)
   - **å›¾è¡¨å†…å®¹ï¼š** è¿™ä¸ªå›¾æ˜¾ç¤ºäº†ä»å‘é€æç¤ºåˆ°æ¥æ”¶æ¨¡å‹å®Œæ•´å“åº”æ‰€éœ€çš„æ—¶é—´ï¼Œå•ä½æ˜¯ç§’ã€‚
   - **è§£é‡Šï¼š** æ—¶å»¶åæ˜ äº†æ¨¡å‹çš„**å“åº”é€Ÿåº¦**ã€‚å¯¹äºéœ€è¦å¿«é€Ÿå“åº”çš„åº”ç”¨ï¼ˆæ¯”å¦‚å®æ—¶èŠå¤©ã€è‡ªåŠ¨åŒ–æµç¨‹ï¼‰ï¼Œæ—¶å»¶æ˜¯ä¸€ä¸ªéå¸¸é‡è¦çš„æŒ‡æ ‡ã€‚é€šå¸¸æç¤ºè¶Šé•¿ã€æ¨¡å‹è¶Šå¤§ï¼Œæ—¶å»¶å¯èƒ½ä¼šè¶Šé«˜ï¼Œå› ä¸ºæ¨¡å‹éœ€è¦å¤„ç†æ›´å¤šçš„ä¿¡æ¯ã€‚ä»å›¾ä¸Šçœ‹ï¼Œ`comprehensive` æç¤ºçš„æ—¶å»¶æ˜¯æœ€é«˜çš„ï¼Œè¿™ç¬¦åˆæˆ‘ä»¬å¯¹é•¿æç¤ºæ—¶å»¶å¯èƒ½å¢åŠ çš„é¢„æœŸã€‚

   ï¼šå¯¹äºæ¯ä¸€ç§æç¤ºç»„åˆï¼Œä»£ç éƒ½è°ƒç”¨å¤§æ¨¡å‹ç”Ÿæˆå“åº”ï¼Œå¹¶æµ‹é‡äº†å‡ ä¸ªå…³é”®æŒ‡æ ‡ï¼š

   - **æç¤º Token**ï¼šå‘é€ç»™æ¨¡å‹çš„æç¤ºçš„é•¿åº¦ã€‚
   - **å“åº” Token**ï¼šæ¨¡å‹ç”Ÿæˆçš„å“åº”çš„é•¿åº¦ã€‚
   - **æ—¶å»¶**ï¼šæ¨¡å‹ç”Ÿæˆå“åº”æ‰€éœ€çš„æ—¶é—´ã€‚
   - **Token æ•ˆç‡**ï¼šå“åº” Token / æç¤º Token çš„æ¯”å€¼ï¼Œç²—ç•¥è¡¡é‡æ¨¡å‹ç”Ÿæˆå†…å®¹çš„æ•ˆç‡ã€‚


```python
def test_layered_contexts(base_prompt: str, context_layers: Dict[str, str]) -> Dict[str, Dict[str, Any]]:
    """
    æµ‹è¯•ä¸åŒä¸Šä¸‹æ–‡å±‚ç»„åˆçš„æ•ˆæœã€‚

    Args:
        base_prompt: åŸºç¡€æç¤ºã€‚
        context_layers: ä¸€ä¸ªå­—å…¸ï¼Œé”®æ˜¯ä¸Šä¸‹æ–‡å±‚çš„åç§°ï¼Œå€¼æ˜¯å¯¹åº”çš„æ–‡æœ¬å†…å®¹ã€‚

    Returns:
        ä¸€ä¸ªå­—å…¸ï¼Œé”®æ˜¯æµ‹è¯•çš„é…ç½®åç§°ï¼ˆe.g., "base", "base+role", "all_layers"ï¼‰ï¼Œ
        å€¼æ˜¯åŒ…å«è¯¥é…ç½®çš„æç¤ºã€å“åº”å’ŒæŒ‡æ ‡çš„å­—å…¸ã€‚
    """
    layer_results: Dict[str, Dict[str, Any]] = {} # å­˜å‚¨æµ‹è¯•ç»“æœ
    print("æµ‹è¯•ï¼šbase")
    # æµ‹è¯•åŸºç¡€æç¤º
    base_resp, base_lat = generate_response(base_prompt)
    base_metrics = calculate_metrics(base_prompt, base_resp, base_lat)
    layer_results["base"] = {
        "prompt": base_prompt,
        "response": base_resp,
        **base_metrics,
    }
    print(f"  æç¤º token: {base_metrics['prompt_tokens']}, å“åº” token: {base_metrics['response_tokens']}, æ—¶å»¶: {base_metrics['latency']:.2f}s, æ•ˆç‡: {base_metrics['token_efficiency']:.2f}")

    # æµ‹è¯•åŸºç¡€æç¤º + å•ä¸ªä¸Šä¸‹æ–‡å±‚
    for layer_name, layer_content in context_layers.items():
        name = f"base+{layer_name}"
        prompt = f"{base_prompt}\n\n{layer_content}" # æ„å»ºç»„åˆæç¤º
        print(f"æµ‹è¯•ï¼š{name}")
        resp, lat = generate_response(prompt)
        metrics = calculate_metrics(prompt, resp, lat)
        layer_results[name] = {
            "prompt": prompt,
            "response": resp,
            **metrics,
        }
        print(f"  æç¤º token: {metrics['prompt_tokens']}, å“åº” token: {metrics['response_tokens']}, æ—¶å»¶: {metrics['latency']:.2f}s, æ•ˆç‡: {metrics['token_efficiency']:.2f}")

    # æµ‹è¯•åŸºç¡€æç¤º + æ‰€æœ‰ä¸Šä¸‹æ–‡å±‚
    all_layers = "\n\n".join(context_layers.values())
    full_prompt = f"{base_prompt}\n\n{all_layers}"
    print("æµ‹è¯•ï¼šall_layers")
    full_resp, full_lat = generate_response(full_prompt)
    full_metrics = calculate_metrics(full_prompt, full_resp, full_lat)
    layer_results["all_layers"] = {
        "prompt": full_prompt,
        "response": full_resp,
        **full_metrics,
    }
    print(f"  æç¤º token: {full_metrics['prompt_tokens']}, å“åº” token: {full_metrics['response_tokens']}, æ—¶å»¶: {full_metrics['latency']:.2f}s, æ•ˆç‡: {full_metrics['token_efficiency']:.2f}")

    return layer_results

# å®šä¹‰ç”¨äºå±‚æµ‹è¯•çš„åŸºç¡€æç¤ºå’Œä¸Šä¸‹æ–‡å±‚
layer_test_prompt = "è¯·ä¸ºæœŸæœ«è€ƒè¯•åå»æˆéƒ½çš„å¤§å­¦ç”Ÿï¼Œè§„åˆ’ä¸€ä¸ªä¸ºæœŸå››å¤©çš„æ—…æ¸¸è·¯çº¿ã€‚"
context_layers = {
    "role": "ä½ çš„èº«ä»½ï¼šä¸€ä½ç†Ÿæ‚‰æˆéƒ½çš„èµ„æ·±æ—…è¡Œåšä¸»ã€‚", # è§’è‰²å±‚
    "requirements": "éœ€æ±‚ï¼š\n- åŒ…å«å¤§ç†ŠçŒ«åŸºåœ°\n- åŒ…å«å®½çª„å··å­æˆ–é”¦é‡Œ\n- åŒ…å«è‡³å°‘ä¸€ç§åœ°é“å·èœä½“éªŒ\n- è¡Œç¨‹èŠ‚å¥è½»æ¾ï¼Œé€‚åˆå­¦ç”Ÿ", # éœ€æ±‚å±‚
    "audience": "ç›®æ ‡å—ä¼—ï¼šæœŸæœ«è€ƒè¯•åæ”¾æ¾çš„å¤§å­¦ç”Ÿï¼Œé¢„ç®—é€‚ä¸­ã€‚", # å—ä¼—å±‚
}
# æ‰§è¡Œå±‚æµ‹è¯•å¹¶è·å–ç»“æœ
layer_test_results = test_layered_contexts(layer_test_prompt, context_layers)

# æ ¹æ® Token æ•ˆç‡æ‰¾åˆ°è¡¨ç°æœ€å¥½çš„é…ç½®
# æ³¨æ„ï¼šè¿™é‡Œä»¥ token_efficiency ä½œä¸ºè¡¡é‡æ ‡å‡†ï¼Œå®é™…åº”ç”¨ä¸­å¯èƒ½éœ€è¦ç»“åˆå…¶ä»–æŒ‡æ ‡ï¼ˆå¦‚è´¨é‡è¯„åˆ†ï¼‰
best = max(layer_test_results.keys(), key=lambda k: layer_test_results[k]['token_efficiency'])
print("\nToken æ•ˆç‡æœ€é«˜é…ç½®ï¼š", best)
```

### 5. ä¸Šä¸‹æ–‡å‹ç¼©ï¼ˆæ‘˜è¦/å…³é”®è¯/è¦ç‚¹ï¼‰

è¿™éƒ¨åˆ†ä»£ç å±•ç¤ºäº†å¦‚ä½•é€šè¿‡å¯¹é•¿ä¸Šä¸‹æ–‡è¿›è¡Œå‹ç¼©æ¥å‡å°‘æç¤ºè¯çš„é•¿åº¦ï¼Œå¹¶å¯¹æ¯”äº†ä½¿ç”¨æ‘˜è¦ã€å…³é”®è¯å’Œè¦ç‚¹ä¸‰ç§å‹ç¼©æŠ€æœ¯åçš„æ•ˆæœã€‚

ä¸åŒçš„å‹ç¼©æ–¹å¼ä¿ç•™ä¿¡æ¯çš„æ–¹å¼ä¸åŒï¼š

- **æ‘˜è¦**ä¿ç•™äº†åŸå§‹ä¸Šä¸‹æ–‡çš„ä¸»è¦æ€æƒ³å’Œæµç•…æ€§ã€‚
- **å…³é”®è¯**æå–äº†æ ¸å¿ƒæ¦‚å¿µï¼Œä½†ä¸¢å¤±äº†åŸæ–‡çš„ç»“æ„å’Œç»†èŠ‚ã€‚
- **è¦ç‚¹**å°†ä¿¡æ¯ç»“æ„åŒ–ï¼Œæœ‰åŠ©äºæ¸…æ™°å‘ˆç°å…³é”®ä¿¡æ¯ã€‚


```python
# å®šä¹‰ä¸€ä¸ªå‡½æ•°ç”¨äºå‹ç¼©ä¸Šä¸‹æ–‡
def compress_context(context: str, technique: str = 'summarize') -> str:
    """
    ä½¿ç”¨ä¸åŒæŠ€æœ¯å‹ç¼©ä¸Šä¸‹æ–‡ã€‚

    Args:
        context: éœ€è¦å‹ç¼©çš„åŸå§‹ä¸Šä¸‹æ–‡ã€‚
        technique: å‹ç¼©æŠ€æœ¯ï¼Œå¯é€‰ 'summarize', 'keywords', 'bullet'ã€‚

    Returns:
        å‹ç¼©åçš„ä¸Šä¸‹æ–‡ã€‚
    """
    # æ ¹æ®æŒ‡å®šçš„å‹ç¼©æŠ€æœ¯æ„å»ºç›¸åº”çš„æç¤º
    if technique == 'summarize':
        prompt = f"è¯·ç²¾ç‚¼ä»¥ä¸‹å…³äºæ—…æ¸¸è·¯çº¿è§„åˆ’çš„ä¸Šä¸‹æ–‡ï¼Œä¿ç•™å…³é”®ä¿¡æ¯å¹¶å‡å°‘å†—ä½™ï¼š\n\n{context}"
    elif technique == 'keywords':
        prompt = f"è¯·æå–ä»¥ä¸‹å…³äºæ—…æ¸¸è·¯çº¿è§„åˆ’çš„ä¸Šä¸‹æ–‡çš„å…³é”®è¯ä¸å…³é”®çŸ­è¯­ï¼ˆé€—å·åˆ†éš”ï¼‰ï¼š\n\n{context}"
    elif technique == 'bullet':
        prompt = f"è¯·å°†ä»¥ä¸‹å…³äºæ—…æ¸¸è·¯çº¿è§„åˆ’çš„ä¸Šä¸‹æ–‡è½¬æ¢ä¸ºç»“æ„åŒ–è¦ç‚¹åˆ—è¡¨ï¼Œè¦†ç›–æ ¸å¿ƒä¿¡æ¯ï¼š\n\n{context}"
    else:
        return context # å¦‚æœæŠ€æœ¯æœªçŸ¥ï¼Œè¿”å›åŸå§‹ä¸Šä¸‹æ–‡
    # è°ƒç”¨æ¨¡å‹ç”Ÿæˆå‹ç¼©åçš„å†…å®¹
    resp, _ = generate_response(prompt)
    return resp

# ä½¿ç”¨ä¹‹å‰å®šä¹‰çš„ comprehensive æç¤ºä½œä¸ºåŸå§‹ä¸Šä¸‹æ–‡è¿›è¡Œå‹ç¼©æµ‹è¯•
original_context = expanded_prompts["comprehensive"]
print("--- åŸå§‹ä¸Šä¸‹æ–‡ ---")
print(original_context)
print("åŸå§‹ tokenï¼š", count_tokens(original_context))

# éå†ä¸åŒçš„å‹ç¼©æŠ€æœ¯è¿›è¡Œæµ‹è¯•
compressed_prompts = {}
for tech in ['summarize', 'keywords', 'bullet']:
    out = compress_context(original_context, tech) # æ‰§è¡Œå‹ç¼©
    compressed_prompts[tech] = out
    print(f"\n--- {tech.upper()} ---") # æ‰“å°æŠ€æœ¯åç§°
    print(out) # æ‰“å°å‹ç¼©åçš„å†…å®¹
    print("å‹ç¼©å tokenï¼š", count_tokens(out)) # æ‰“å°å‹ç¼©åçš„ Token æ•°é‡

print("\n" + "="*50)
print("--- ç”Ÿæˆå†…å®¹å¯¹æ¯” ---")

# ç”ŸæˆåŸå§‹æç¤ºè¯çš„å“åº”
print("\n--- åŸå§‹æç¤ºè¯ç”Ÿæˆå†…å®¹ ---")
original_response, original_latency = generate_response(original_context)
original_metrics = calculate_metrics(original_context, original_response, original_latency)
print(original_response)
print(f"  æç¤º token: {original_metrics['prompt_tokens']}, å“åº” token: {original_metrics['response_tokens']}, æ—¶å»¶: {original_metrics['latency']:.2f}s, æ•ˆç‡: {original_metrics['token_efficiency']:.2f}")


# ç”Ÿæˆå‹ç¼©åæç¤ºè¯çš„å“åº”
for tech, compressed_prompt in compressed_prompts.items():
    print(f"\n--- {tech.upper()} å‹ç¼©åæç¤ºè¯ç”Ÿæˆå†…å®¹ ---")
    compressed_response, compressed_latency = generate_response(compressed_prompt)
    compressed_metrics = calculate_metrics(compressed_prompt, compressed_response, compressed_latency)
    print(compressed_response)
    print(f"  æç¤º token: {compressed_metrics['prompt_tokens']}, å“åº” token: {compressed_metrics['response_tokens']}, æ—¶å»¶: {compressed_metrics['latency']:.2f}s, æ•ˆç‡: {compressed_metrics['token_efficiency']:.2f}")
```

### 6. ä¸Šä¸‹æ–‡ä¿®å‰ªï¼šç³»ç»ŸåŒ–åˆ é™¤æ— æ•ˆå±‚ï¼ˆå¯é€‰ï¼‰

ä½¿ç”¨æ¨¡å‹è‡ªè¯„æ¥æ¯”è¾ƒä¸åŒå±‚ç»„åˆä¸‹çš„è´¨é‡åˆ†ï¼Œå¹¶é€’å½’åˆ é™¤æ— æ•ˆå±‚ã€‚

å±•ç¤ºä¸€ç§**æ›´ç³»ç»Ÿã€æ›´æ•°æ®é©±åŠ¨**çš„æ–¹æ³•æ¥ä¼˜åŒ–æç¤ºè¯ä¸­çš„ä¸Šä¸‹æ–‡å±‚ã€‚

å®ƒçš„æ ¸å¿ƒé€»è¾‘æ˜¯ï¼š

1. **å®šä¹‰è¯„ä¼°æ ‡å‡†**ï¼šé¦–å…ˆï¼Œæ˜ç¡®ä½ å¸Œæœ›æ¨¡å‹ç”Ÿæˆçš„å“åº”åœ¨å“ªäº›æ–¹é¢æ˜¯â€œå¥½â€çš„ï¼ˆä¾‹å¦‚ï¼Œæ˜¯å¦åŒ…å«æ‰€æœ‰è¦æ±‚çš„ä¿¡æ¯ã€è¯­æ°”æ˜¯å¦åˆé€‚ã€ç»“æ„æ˜¯å¦æ¸…æ™°ç­‰ï¼‰ã€‚
2. **æ¨¡å‹è‡ªè¯„**ï¼šåˆ©ç”¨æ¨¡å‹æœ¬èº«çš„è¯„ä¼°èƒ½åŠ›ï¼Œè®©å®ƒæ ¹æ®ä½ å®šä¹‰çš„æ ‡å‡†ç»™ä¸åŒæç¤ºè¯ç»„åˆç”Ÿæˆçš„å“åº”æ‰“åˆ†ã€‚
3. **ç³»ç»Ÿæ€§æµ‹è¯•å’Œä¿®å‰ª**ï¼šä»£ç ä¼šæµ‹è¯•åŒ…å«ä¸åŒä¸Šä¸‹æ–‡å±‚ç»„åˆçš„æç¤ºè¯ï¼ˆä¾‹å¦‚ï¼ŒåªåŒ…å«åŸºç¡€æç¤ºã€åŒ…å«åŸºç¡€+è§’è‰²ã€åŒ…å«åŸºç¡€+è§’è‰²+çº¦æŸç­‰ï¼‰ï¼Œå¹¶è¯„ä¼°å®ƒä»¬çš„è´¨é‡å¾—åˆ†å’Œ Token æ•°é‡ã€‚
4. **é€’å½’ä¼˜åŒ–ï¼ˆæ¦‚å¿µï¼‰**ï¼šç†è®ºä¸Šï¼Œä½ å¯ä»¥é€šè¿‡æ¯”è¾ƒä¸åŒç»„åˆçš„æ•ˆæœï¼Œè¯†åˆ«å‡ºé‚£äº›å¢åŠ äº†æç¤ºè¯é•¿åº¦ï¼ˆToken æ•°é‡ï¼‰ä½†å¯¹æé«˜å“åº”è´¨é‡å¸®åŠ©ä¸å¤§çš„ä¸Šä¸‹æ–‡å±‚ï¼Œç„¶åå°†å®ƒä»¬ç§»é™¤ï¼Œç”šè‡³è¿›è¡Œé€’å½’çš„ä¿®å‰ªè¿‡ç¨‹ï¼Œç›´åˆ°æ‰¾åˆ°ä¸€ä¸ªåœ¨è´¨é‡å’Œæˆæœ¬ä¹‹é—´è¾¾åˆ°æœ€ä½³å¹³è¡¡çš„æç¤ºè¯ç»“æ„ã€‚


```python
import re

# å®šä¹‰ä¸€ä¸ªå‡½æ•°ï¼Œè®©æ¨¡å‹è¯„ä¼°ç”Ÿæˆå“åº”çš„è´¨é‡
def evaluate_response_quality(prompt: str, response: str, criteria: List[str]) -> float:
    """
    ä½¿ç”¨æ¨¡å‹è¯„ä¼°å“åº”è´¨é‡å¹¶æå–æ€»ä½“è¯„åˆ†ã€‚

    Args:
        prompt: ç”¨äºç”Ÿæˆå“åº”çš„æç¤ºã€‚
        response: éœ€è¦è¯„ä¼°çš„å“åº”ã€‚
        criteria: ç”¨äºè¯„ä¼°å“åº”çš„è´¨é‡æ ‡å‡†åˆ—è¡¨ã€‚

    Returns:
        æ€»ä½“è¯„åˆ† (0.0-1.0)ï¼Œå¦‚æœæ— æ³•æå–è¯„åˆ†åˆ™è¿”å› 0.5ã€‚
    """
    # æ„å»ºç”¨äºè¯„ä¼°çš„æç¤ºï¼ŒåŒ…å«åŸå§‹æç¤ºã€å“åº”å’Œè¯„ä¼°æ ‡å‡†
    crit = "\n".join([f"- {c}" for c in criteria]) # æ ¼å¼åŒ–è¯„ä¼°æ ‡å‡†
    eval_prompt = (
        "è¯·ä¸ºä»¥ä¸‹å…³äºæ—…æ¸¸è·¯çº¿è§„åˆ’çš„å“åº”åŸºäºæ ‡å‡†æ‰“åˆ†ï¼Œæœ€åç»™å‡ºâ€˜æ€»ä½“è¯„åˆ†ï¼š0.0-1.0â€™ã€‚\n\n"
        f"æç¤ºï¼š\n{prompt}\n\nå“åº”ï¼š\n{response}\n\nè¯„ä¼°æ ‡å‡†ï¼š\n{crit}\n"
    )
    # è°ƒç”¨æ¨¡å‹ç”Ÿæˆè¯„ä¼°ç»“æœ
    eval_resp, _ = generate_response(eval_prompt)
    # ä½¿ç”¨æ­£åˆ™è¡¨è¾¾å¼ä»è¯„ä¼°ç»“æœä¸­æå–æ€»ä½“è¯„åˆ†
    m = re.findall(r"æ€»ä½“è¯„åˆ†[:ï¼š]\s*([0-9]*\.?[0-9]+)", eval_resp)
    # å¦‚æœæå–åˆ°è¯„åˆ†ï¼Œè¿”å›æœ€åä¸€ä¸ªè¯„åˆ†ï¼ˆå¤„ç†å¯èƒ½æœ‰å¤šä¸ªè¯„åˆ†çš„æƒ…å†µï¼‰ï¼Œå¦åˆ™è¿”å› 0.5
    return float(m[-1]) if m else 0.5

# å®šä¹‰ä¸€ä¸ªå‡½æ•°ï¼Œé€šè¿‡é€’å½’æ–¹å¼ä¿®å‰ªä¸Šä¸‹æ–‡å±‚
def prune_context_layers(base_prompt: str, layers: Dict[str, str], criteria: List[str]) -> Tuple[str, Dict[str, Any]]:
    """
    ç³»ç»Ÿæ€§åœ°ä¿®å‰ªä¸Šä¸‹æ–‡å±‚ï¼Œä¿ç•™å¯¹è´¨é‡æå‡æœ‰ç›Šçš„å±‚ã€‚

    Args:
        base_prompt: åŸºç¡€æç¤ºã€‚
        layers: éœ€è¦æµ‹è¯•å’Œä¿®å‰ªçš„ä¸Šä¸‹æ–‡å±‚å­—å…¸ã€‚
        criteria: ç”¨äºè¯„ä¼°å“åº”è´¨é‡çš„æ ‡å‡†åˆ—è¡¨ã€‚

    Returns:
        æœ€ä¼˜çš„æç¤ºå­—ç¬¦ä¸²å’ŒåŒ…å«æ‰€æœ‰æµ‹è¯•ç»“æœçš„å­—å…¸ã€‚
    """
    print("--- å¼€å§‹ä¸Šä¸‹æ–‡å±‚ä¿®å‰ªæµ‹è¯• ---")
    results: Dict[str, Any] = {} # å­˜å‚¨æµ‹è¯•ç»“æœ

    # 1. æµ‹è¯•åŸºç¡€æç¤ºçš„è´¨é‡
    print("æµ‹è¯•ï¼šbase")
    base_resp, _ = generate_response(base_prompt)
    base_quality = evaluate_response_quality(base_prompt, base_resp, criteria)
    base_tokens = count_tokens(base_prompt)
    results["base"] = {"prompt": base_prompt, "quality": base_quality, "tokens": base_tokens}
    print(f"  åŸºç¡€æç¤ºè´¨é‡ï¼š{base_quality:.2f}, Token: {base_tokens}")

    # å¦‚æœæ²¡æœ‰ä¸Šä¸‹æ–‡å±‚éœ€è¦æµ‹è¯•ï¼Œç›´æ¥è¿”å›åŸºç¡€æç¤ºç»“æœ
    if not layers:
        print("--- æ²¡æœ‰å¯ä¿®å‰ªçš„ä¸Šä¸‹æ–‡å±‚ ---")
        return base_prompt, results

    # 2. æµ‹è¯•åŒ…å«æ‰€æœ‰ä¸Šä¸‹æ–‡å±‚çš„æç¤ºçš„è´¨é‡
    all_text = "\n\n".join(layers.values())
    full_prompt = f"{base_prompt}\n\n{all_text}"
    print("æµ‹è¯•ï¼šall_layers")
    full_resp, _ = generate_response(full_prompt)
    full_quality = evaluate_response_quality(full_prompt, full_resp, criteria)
    full_tokens = count_tokens(full_prompt)
    results["all_layers"] = {"prompt": full_prompt, "quality": full_quality, "tokens": full_tokens}
    print(f"  æ‰€æœ‰å±‚ç»„åˆè´¨é‡ï¼š{full_quality:.2f}, Token: {full_tokens}")

    best_quality = full_quality # åˆå§‹åŒ–æœ€ä½³è´¨é‡ä¸ºæ‰€æœ‰å±‚ç»„åˆçš„è´¨é‡
    best_key = "all_layers" # åˆå§‹åŒ–æœ€ä½³é…ç½®ä¸ºæ‰€æœ‰å±‚ç»„åˆ

    # 3. é€ä¸ªç§»é™¤ä¸Šä¸‹æ–‡å±‚å¹¶æµ‹è¯•å‰©ä½™ç»„åˆçš„è´¨é‡
    current_layers = layers.copy() # å¤åˆ¶å½“å‰ä¸Šä¸‹æ–‡å±‚å­—å…¸
    removed_layer = None # è®°å½•å½“å‰è½®æ¬¡ç§»é™¤çš„å±‚

    # å°è¯•ç§»é™¤æ¯ä¸ªå±‚ï¼Œçœ‹çœ‹æ˜¯å¦èƒ½æé«˜æˆ–ä¿æŒè´¨é‡åŒæ—¶å‡å°‘ Token
    for name in list(current_layers.keys()):
        remain = {k: v for k, v in current_layers.items() if k != name} # ç§»é™¤å½“å‰å±‚åçš„å‰©ä½™å±‚
        test_prompt = f"{base_prompt}\n\n" + "\n\n".join(remain.values()) # æ„å»ºæµ‹è¯•æç¤º
        print(f"æµ‹è¯•ï¼šwithout_{name}")
        test_resp, _ = generate_response(test_prompt)
        q = evaluate_response_quality(test_prompt, test_resp, criteria) # è¯„ä¼°è´¨é‡
        key = f"without_{name}"
        test_tokens = count_tokens(test_prompt)
        results[key] = {"prompt": test_prompt, "quality": q, "tokens": test_tokens}
        print(f"  ç§»é™¤ {name} åè´¨é‡ï¼š{q:.2f}, Token: {test_tokens}")

        # åˆ¤æ–­å½“å‰ç»„åˆæ˜¯å¦æ˜¯æ›´å¥½çš„é€‰æ‹©ï¼š
        # å¦‚æœè´¨é‡æ˜¾è‘—æå‡ï¼ˆä¾‹å¦‚ > 0.05ï¼‰ï¼Œæˆ–è€…è´¨é‡ç›¸å½“ä½† Token å‡å°‘ï¼Œåˆ™æ›´æ–°æœ€ä½³é…ç½®
        if q > best_quality + 0.05: # è´¨é‡æ˜¾è‘—æå‡
             best_quality, best_key = q, key
             removed_layer = name
        elif abs(q - best_quality) <= 0.05 and test_tokens < results[best_key]["tokens"]: # è´¨é‡ç›¸å½“ä½† Token å‡å°‘
             best_quality, best_key = q, key
             removed_layer = name

    print(f"æœ¬è½®æœ€ä½³é…ç½®ï¼š{best_key}, è´¨é‡: {best_quality:.2f}, Token: {results[best_key]['tokens']}")

    # 4. é€’å½’è°ƒç”¨ï¼šå¦‚æœæœ¬è½®ç§»é™¤äº†å±‚å¹¶æ‰¾åˆ°äº†æ›´å¥½çš„é…ç½®ï¼Œåˆ™åŸºäºå‰©ä½™å±‚ç»§ç»­ä¿®å‰ª
    if removed_layer:
        print(f"å‘ç°å¯ç§»é™¤å±‚ï¼š{removed_layer}ï¼Œç»§ç»­ä¿®å‰ª...")
        new_layers = {k: v for k, v in current_layers.items() if k != removed_layer}
        # å°†å½“å‰è½®æ¬¡çš„ç»“æœåˆå¹¶åˆ°ä¸‹ä¸€è½®çš„ results ä¸­
        next_prompt, next_results = prune_context_layers(base_prompt, new_layers, criteria)
        results.update(next_results) # åˆå¹¶ç»“æœ
        return next_prompt, results
    else:
        # å¦‚æœæœ¬è½®æ²¡æœ‰ç§»é™¤ä»»ä½•å±‚ï¼Œè¯´æ˜å·²ç»æ‰¾åˆ°å±€éƒ¨æœ€ä¼˜è§£æˆ–æ— æ³•è¿›ä¸€æ­¥ä¼˜åŒ–
        print("--- ä¿®å‰ªå®Œæˆï¼Œæœªå‘ç°æ›´å¤šå¯ç§»é™¤å±‚ ---")
        return results[best_key]["prompt"], results # è¿”å›æœ€ä½³æç¤ºå’Œæ‰€æœ‰æµ‹è¯•ç»“æœ

# ç¤ºä¾‹ç”¨æ³•ï¼šå®šä¹‰åŸºç¡€æç¤ºã€ä¸Šä¸‹æ–‡å±‚å’Œè¯„ä¼°æ ‡å‡†
prune_test_prompt = "è¯·ä¸ºæœŸæœ«è€ƒè¯•åå»åŒ—äº¬çš„å¤§å­¦ç”Ÿï¼Œè§„åˆ’ä¸€ä¸ªä¸ºæœŸäº”å¤©çš„æ—…æ¸¸è·¯çº¿ã€‚"
prune_context_layers_dict = {
    "role": "ä½ çš„èº«ä»½ï¼šä¸€ä½ç†Ÿæ‚‰åŒ—äº¬çš„èµ„æ·±æ—…è¡Œåšä¸»ã€‚",
    "audience": "ç›®æ ‡è¯»è€…ï¼šæœŸæœ«è€ƒè¯•åæ”¾æ¾çš„å¤§å­¦ç”Ÿï¼Œå¯¹å†å²æ–‡åŒ–å’Œç°ä»£éƒ½å¸‚éƒ½æ„Ÿå…´è¶£ã€‚",
    "constraints": "è¦æ±‚ï¼š\n- åŒ…å«æ•…å®«å’Œé•¿åŸ\n- åŒ…å«è‡³å°‘ä¸€ä¸ªè‰ºæœ¯åŒºæˆ–åˆ›æ„å›­åŒº\n- åŒ…å«è‡³å°‘ä¸€ç§åŒ—äº¬ç‰¹è‰²å°åƒä½“éªŒ\n- è€ƒè™‘å…¬å…±äº¤é€šä¾¿æ·æ€§\n- è¡Œç¨‹å®‰æ’åŠ³é€¸ç»“åˆ",
    "example": "ç¤ºä¾‹ï¼š\nç¬¬ä¸€å¤©ï¼šä¸ŠåˆæŠµè¾¾åŒ—äº¬ï¼Œå‰å¾€é…’åº—ã€‚ä¸‹åˆå‚è§‚æ•…å®«ï¼Œæ„Ÿå—çš‡å®¶æ°”æ´¾ã€‚æ™šä¸Šå¯ä»¥åœ¨ç‹åºœäº•å¤§è¡—å“å°åŒ—äº¬çƒ¤é¸­ã€‚\nç¬¬äºŒå¤©ï¼šä¸Šåˆå‰å¾€æ…•ç”°å³ªé•¿åŸï¼Œä½“éªŒå£®ä¸½æ™¯è‰²ã€‚ä¸‹åˆè¿”å›å¸‚åŒºï¼Œå‚è§‚798è‰ºæœ¯åŒºã€‚",
}
prune_criteria = [
    "è¡Œç¨‹è§„åˆ’æ˜¯å¦åˆç†ï¼Œè€ƒè™‘äº†äº¤é€šå’ŒèŠ‚å¥ï¼Ÿ",
    "æ˜¯å¦åŒ…å«äº†æ•…å®«å’Œé•¿åŸï¼Ÿ",
    "æ˜¯å¦åŒ…å«äº†è‡³å°‘ä¸€ä¸ªè‰ºæœ¯åŒºæˆ–åˆ›æ„å›­åŒºï¼Ÿ",
    "æ˜¯å¦åŒ…å«äº†è‡³å°‘ä¸€ç§åŒ—äº¬ç‰¹è‰²å°åƒä½“éªŒï¼Ÿ",
    "æ˜¯å¦é€‚åˆæœŸæœ«è€ƒè¯•åæ”¾æ¾çš„å¤§å­¦ç”Ÿï¼Ÿ",
    "ä½œä¸ºæ—…è¡Œåšä¸»ï¼Œè¯­æ°”æ˜¯å¦ä¸“ä¸šä¸”æœ‰å¸å¼•åŠ›ï¼Ÿ",
]

# æ‰§è¡Œä¸Šä¸‹æ–‡å±‚ä¿®å‰ªè¿‡ç¨‹
pruned_prompt, prune_results = prune_context_layers(prune_test_prompt, prune_context_layers_dict, prune_criteria)
print("\n--- ä¿®å‰ªç»“æœ ---")
print("æœ€ä¼˜æç¤ºï¼š\n", pruned_prompt)
print("\nå„é…ç½®æµ‹è¯•ç»“æœï¼š")
for key, res in prune_results.items():
   print(f"  {key}: è´¨é‡={res['quality']:.2f}, Token={res['tokens']}")

# æ³¨æ„ï¼šæ¨¡å‹è‡ªè¯„å¯èƒ½å¼•å…¥ä¸ç¨³å®šæ€§
```

- **å¹¶éæ‰€æœ‰ä¸Šä¸‹æ–‡éƒ½æœ‰ç›Š**ï¼šå¢åŠ ä¸Šä¸‹æ–‡å¹¶ä¸æ€»æ˜¯èƒ½æå‡æ¨¡å‹è¡¨ç°ã€‚æœ‰äº›ä¿¡æ¯å¯èƒ½å†—ä½™ï¼Œç”šè‡³ä¼šå¹²æ‰°æ¨¡å‹ç†è§£æ ¸å¿ƒä»»åŠ¡ã€‚
- **éœ€è¦ç³»ç»Ÿæ€§æµ‹è¯•æ¥æ‰¾åˆ°æœ€ä¼˜è§£**ï¼šé€šè¿‡æµ‹è¯•ä¸åŒçš„ä¸Šä¸‹æ–‡å±‚ç»„åˆï¼Œæ‰èƒ½æ‰¾åˆ°åœ¨ä½ çš„ç‰¹å®šä»»åŠ¡ä¸­æ—¢èƒ½ä¿è¯è´¨é‡åˆèƒ½æ§åˆ¶æˆæœ¬ï¼ˆTokenï¼‰çš„æœ€ä½³æç¤ºè¯ç»“æ„ã€‚
- **æ¨¡å‹è‡ªè¯„æ˜¯ä¸€ç§æ½œåœ¨å·¥å…·**ï¼šè™½ç„¶æ¨¡å‹è‡ªè¯„å¯èƒ½å¼•å…¥ä¸ç¨³å®šæ€§ï¼ˆæ¨¡å‹è¯„ä¼°è‡ªå·±å¯èƒ½ä¼šæœ‰åè§æˆ–ä¸ä¸€è‡´ï¼‰ï¼Œä½†å®ƒæä¾›äº†ä¸€ç§è‡ªåŠ¨åŒ–çš„æ–¹å¼æ¥åˆæ­¥è¯„ä¼°ä¸åŒæç¤ºè¯çš„æ•ˆæœï¼Œå¯ä»¥ä½œä¸ºäººå·¥è¯„ä¼°çš„è¡¥å……æˆ–åˆæ­¥ç­›é€‰å·¥å…·ã€‚
- **â€œä¿®å‰ªâ€çš„é‡è¦æ€§**ï¼šå­¦ä¼šå¦‚ä½•è¯†åˆ«å’Œç§»é™¤æç¤ºè¯ä¸­â€œæ— æ•ˆâ€æˆ–ä½æ•ˆçš„ä¸Šä¸‹æ–‡å±‚ï¼Œæ˜¯æ§åˆ¶æˆæœ¬å’Œæå‡æ¨¡å‹å¤„ç†æ•ˆç‡çš„å…³é”®æŠ€å·§ã€‚

### 7. ç»“åˆæ£€ç´¢ï¼ˆRAGï¼‰

ç”¨ç®€å•å…³é”®è¯åŒ¹é…æ¼”ç¤ºä»â€œçŸ¥è¯†åº“â€å–å›ä¿¡æ¯å¹¶æ‹¼æ¥åˆ°æç¤ºä¸­ã€‚

è¿™éƒ¨åˆ†ä»£ç çš„æ ¸å¿ƒæ˜¯å¼•å…¥äº† **æ£€ç´¢å¢å¼ºç”Ÿæˆï¼ˆRetrieval Augmented Generation, RAGï¼‰** çš„æ¦‚å¿µã€‚å®ƒçš„åŸºæœ¬æ€æƒ³æ˜¯ï¼šå½“æ¨¡å‹éœ€è¦å›ç­”ä¸€ä¸ªé—®é¢˜æˆ–å®Œæˆä¸€ä¸ªä»»åŠ¡æ—¶ï¼Œä¸æ˜¯ä»…ä»…ä¾é å…¶å†…ç½®çš„çŸ¥è¯†ï¼Œè€Œæ˜¯å…ˆä»ä¸€ä¸ªå¤–éƒ¨çš„çŸ¥è¯†åº“ä¸­**æ£€ç´¢**å‡ºä¸å½“å‰ä»»åŠ¡æˆ–é—®é¢˜ç›¸å…³çš„**ä¿¡æ¯**ï¼Œç„¶åå°†è¿™äº›æ£€ç´¢åˆ°çš„ä¿¡æ¯ä½œä¸ºé¢å¤–çš„**ä¸Šä¸‹æ–‡**æ·»åŠ åˆ°æç¤ºè¯ä¸­ï¼Œæœ€åè®©æ¨¡å‹åŸºäºè¿™ä¸ªåŒ…å«å¤–éƒ¨ä¿¡æ¯çš„æç¤ºè¯æ¥**ç”Ÿæˆ**å“åº”ã€‚

ä»£ç ä¸­ç”¨ä¸€ä¸ªç®€åŒ–çš„ä¾‹å­æ¼”ç¤ºäº†è¿™ä¸ªè¿‡ç¨‹ï¼š

1. **æ¨¡æ‹ŸçŸ¥è¯†åº“**ï¼šåˆ›å»ºäº†ä¸€ä¸ªåŒ…å«ä¸€äº›æ—…æ¸¸è§„åˆ’ç›¸å…³åŸºç¡€çŸ¥è¯†çš„ç®€å•åˆ—è¡¨ä½œä¸ºâ€œçŸ¥è¯†åº“â€ã€‚
2. **æ¨¡æ‹Ÿæ£€ç´¢**ï¼šå®šä¹‰äº†ä¸€ä¸ª `retrieve_relevant_info` å‡½æ•°ï¼Œå®ƒæ ¹æ®è¾“å…¥çš„æŸ¥è¯¢ï¼ˆæ¯”å¦‚â€œæ—…æ¸¸è§„åˆ’ è‡ªç”±è¡Œ æ­¥éª¤â€ï¼‰åœ¨è¿™ä¸ªæ¨¡æ‹ŸçŸ¥è¯†åº“ä¸­è¿›è¡Œç®€å•çš„å…³é”®è¯åŒ¹é…ï¼Œæ‰¾å‡ºç›¸å…³çš„çŸ¥è¯†æ¡ç›®ã€‚
3. **æ„å»º RAG ä¸Šä¸‹æ–‡**ï¼šå®šä¹‰äº†ä¸€ä¸ª `create_rag_context` å‡½æ•°ï¼Œå®ƒå°†åŸå§‹çš„åŸºç¡€æç¤ºå’Œæ£€ç´¢åˆ°çš„ç›¸å…³ä¿¡æ¯ç»„åˆèµ·æ¥ï¼Œå½¢æˆä¸€ä¸ªåŒ…å«å¤–éƒ¨çŸ¥è¯†çš„â€œRAG ä¸Šä¸‹æ–‡â€ã€‚
4. **è°ƒç”¨æ¨¡å‹ç”Ÿæˆå“åº”**ï¼šä½¿ç”¨è¿™ä¸ªåŒ…å«æ£€ç´¢ä¿¡æ¯çš„ RAG ä¸Šä¸‹æ–‡å»è°ƒç”¨å¤§æ¨¡å‹ï¼Œå¹¶æ‰“å°æ¨¡å‹ç”Ÿæˆçš„å“åº”ã€‚

```python
from typing import Dict as TDict

# æ¨¡æ‹Ÿä¸€ä¸ªç®€å•çš„çŸ¥è¯†åº“ï¼ŒåŒ…å«ä¸€äº›å…³äºæ—…æ¸¸è§„åˆ’çš„ä¿¡æ¯
kb = [
    {"title": "æ—…æ¸¸è§„åˆ’åŸºç¡€", "content": "æ—…æ¸¸è§„åˆ’æ˜¯åˆ¶å®šæ—…è¡Œè¡Œç¨‹çš„è¿‡ç¨‹ï¼ŒåŒ…æ‹¬é€‰æ‹©ç›®çš„åœ°ã€äº¤é€šæ–¹å¼ã€ä½å®¿ã€æ™¯ç‚¹å’Œæ´»åŠ¨ç­‰ã€‚"},
    {"title": "é€‰æ‹©ç›®çš„åœ°", "content": "é€‰æ‹©ç›®çš„åœ°éœ€è¦è€ƒè™‘æ—…è¡Œæ—¶é—´ã€é¢„ç®—ã€æ—…è¡Œåå¥½ï¼ˆå¦‚æ–‡åŒ–ã€è‡ªç„¶ã€æ¢é™©ï¼‰ä»¥åŠåŒè¡Œäººå‘˜ã€‚"},
    {"title": "äº¤é€šæ–¹å¼é€‰æ‹©", "content": "å¸¸è§çš„äº¤é€šæ–¹å¼åŒ…æ‹¬é£æœºã€ç«è½¦ã€æ±½è½¦å’Œè½®èˆ¹ã€‚é€‰æ‹©å“ªç§æ–¹å¼å–å†³äºç›®çš„åœ°è·ç¦»ã€é¢„ç®—å’Œæ—¶é—´ã€‚"},
    {"title": "ä½å®¿é¢„è®¢", "content": "ä½å®¿ç±»å‹å¤šæ ·ï¼Œå¦‚é…’åº—ã€æ°‘å®¿ã€é’å¹´æ—…ç¤¾ç­‰ã€‚é¢„è®¢æ—¶éœ€è€ƒè™‘ä½ç½®ã€ä»·æ ¼ã€è®¾æ–½å’Œè¯„ä»·ã€‚"},
    {"title": "æ™¯ç‚¹å’Œæ´»åŠ¨å®‰æ’", "content": "æ ¹æ®å…´è¶£å’Œæ—¶é—´å®‰æ’å‚è§‚æ™¯ç‚¹å’Œå‚ä¸æ´»åŠ¨ï¼Œæ³¨æ„å¼€æ”¾æ—¶é—´ã€é—¨ç¥¨ä¿¡æ¯å’Œé¢„çº¦éœ€æ±‚ã€‚"},
    {"title": "é¢„ç®—æ§åˆ¶", "content": "æå‰åˆ¶å®šé¢„ç®—ï¼ŒåŒ…æ‹¬äº¤é€šã€ä½å®¿ã€é¤é¥®ã€é—¨ç¥¨ã€è´­ç‰©ç­‰å„é¡¹å¼€æ”¯ï¼Œå¹¶åœ¨æ—…è¡Œä¸­å°½é‡éµå®ˆã€‚"},
    {"title": "æ‰“åŒ…è¡Œæ", "content": "æ ¹æ®ç›®çš„åœ°æ°”å€™ã€æ—…è¡Œå¤©æ•°å’Œæ´»åŠ¨å†…å®¹å‡†å¤‡è¡£ç‰©ã€æ´—æ¼±ç”¨å“ã€è¯å“å’Œå¿…è¦è¯ä»¶ã€‚"},
    {"title": "å®‰å…¨æ³¨æ„äº‹é¡¹", "content": "äº†è§£ç›®çš„åœ°å®‰å…¨æƒ…å†µï¼Œä¿ç®¡å¥½ä¸ªäººè´¢ç‰©ï¼Œéµå®ˆå½“åœ°æ³•å¾‹æ³•è§„ï¼Œè´­ä¹°æ—…è¡Œä¿é™©ã€‚"},
]

# å®šä¹‰ä¸€ä¸ªå‡½æ•°ï¼Œæ ¹æ®ç”¨æˆ·æŸ¥è¯¢ä»çŸ¥è¯†åº“ä¸­æ£€ç´¢ç›¸å…³ä¿¡æ¯
def retrieve_relevant_info(query: str, knowledge_base: List[TDict[str, str]]) -> List[str]:
    """
    æ ¹æ®æŸ¥è¯¢ä»çŸ¥è¯†åº“ä¸­æ£€ç´¢ç›¸å…³ä¿¡æ¯ã€‚

    Args:
        query: ç”¨æˆ·æŸ¥è¯¢å­—ç¬¦ä¸²ã€‚
        knowledge_base: çŸ¥è¯†åº“ï¼ŒåŒ…å«å­—å…¸åˆ—è¡¨ï¼Œæ¯ä¸ªå­—å…¸æœ‰ 'title' å’Œ 'content' é”®ã€‚

    Returns:
        ä¸€ä¸ªåŒ…å«æ£€ç´¢åˆ°çš„ç›¸å…³ä¿¡æ¯å†…å®¹å­—ç¬¦ä¸²çš„åˆ—è¡¨ï¼ŒæŒ‰ç›¸å…³åº¦é™åºæ’åˆ—ï¼ˆæœ€å¤šè¿”å› 3 æ¡ï¼‰ã€‚
    """
    q_terms = set(query.lower().split()) # å°†æŸ¥è¯¢è½¬æ¢ä¸ºå°å†™å•è¯é›†åˆ
    hits: List[Tuple[int, str]] = [] # å­˜å‚¨åŒ¹é…åˆ°çš„ä¿¡æ¯å’Œå¾—åˆ†
    for item in knowledge_base:
        content = item['content'].lower() # çŸ¥è¯†åº“å†…å®¹è½¬å°å†™
        title = item['title'].lower() # çŸ¥è¯†åº“æ ‡é¢˜è½¬å°å†™
        # è®¡ç®—æŸ¥è¯¢è¯åœ¨å†…å®¹æˆ–æ ‡é¢˜ä¸­å‡ºç°çš„æ¬¡æ•°ä½œä¸ºå¾—åˆ†
        score = sum(1 for t in q_terms if t in content or t in title)
        if score > 0: # åªæ·»åŠ æœ‰åŒ¹é…å¾—åˆ†çš„ä¿¡æ¯
            hits.append((score, item['content']))
    hits.sort(key=lambda x: x[0], reverse=True) # æŒ‰å¾—åˆ†é™åºæ’åº
    return [h[1] for h in hits[:3]] # è¿”å›å¾—åˆ†æœ€é«˜çš„ 3 æ¡ä¿¡æ¯å†…å®¹

# å®šä¹‰ä¸€ä¸ªå‡½æ•°ï¼Œå°†æ£€ç´¢åˆ°çš„ä¿¡æ¯æ·»åŠ åˆ°åŸºç¡€æç¤ºä¸­æ„å»º RAG ä¸Šä¸‹æ–‡
def create_rag_context(base_prompt: str, query: str, kb_list: List[TDict[str, str]]) -> str:
    """
    åˆ›å»ºç»“åˆæ£€ç´¢å¢å¼ºçš„ä¸Šä¸‹æ–‡ã€‚

    Args:
        base_prompt: åŸºç¡€æç¤ºã€‚
        query: ç”¨äºæ£€ç´¢çš„æŸ¥è¯¢ã€‚
        kb_list: çŸ¥è¯†åº“åˆ—è¡¨ã€‚

    Returns:
        ç»“åˆäº†æ£€ç´¢ä¿¡æ¯çš„æ‰©å±•æç¤ºã€‚
    """
    infos = retrieve_relevant_info(query, kb_list) # æ£€ç´¢ç›¸å…³ä¿¡æ¯
    if not infos:
        return base_prompt # å¦‚æœæ²¡æœ‰æ£€ç´¢åˆ°ä¿¡æ¯ï¼Œè¿”å›åŸºç¡€æç¤º
    # å¦‚æœæ£€ç´¢åˆ°ä¿¡æ¯ï¼Œå°†ä¿¡æ¯æ·»åŠ åˆ°æç¤ºåé¢
    return base_prompt + "\n\nç›¸å…³ä¿¡æ¯ï¼š\n\n" + "\n\n".join(infos)

# å®šä¹‰ä¸€ä¸ªç”¨äº RAG æµ‹è¯•çš„æç¤ºå’ŒæŸ¥è¯¢
rag_prompt = "è¯·ç®€è¦è®²è§£å¦‚ä½•è§„åˆ’ä¸€æ¬¡è‡ªç”±è¡Œã€‚"
rag_query = "æ—…æ¸¸è§„åˆ’ è‡ªç”±è¡Œ æ­¥éª¤" # ç”¨äºæ£€ç´¢çš„å…³é”®è¯
# åˆ›å»º RAG ä¸Šä¸‹æ–‡
rag_ctx = create_rag_context(rag_prompt, rag_query, kb)
# æ‰“å° RAG ä¸Šä¸‹æ–‡
print("--- RAG ä¸Šä¸‹æ–‡ ---")
print(rag_ctx)
# ä½¿ç”¨ RAG ä¸Šä¸‹æ–‡è°ƒç”¨æ¨¡å‹ç”Ÿæˆå“åº”
rag_resp, _ = generate_response(rag_ctx)
# æ‰“å° RAG å“åº”
print("\n--- RAG å“åº” ---")
print(rag_resp)
```
